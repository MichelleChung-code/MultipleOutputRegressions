```
from pyspark.ml import Pipeline
from pyspark.ml.classification import RandomForestClassifier, GBTClassifier, DecisionTreeClassifier
```

```
data = sqlContext.sql('select * from sample_libsvm_data')

# data from https://github.com/apache/spark/blob/master/data/mllib/sample_libsvm_data.txt
```

```
train_data, test_data = data.randomSplit([0.7, 0.3])
```

```
dt_classifier = DecisionTreeClassifier()
rf_classifier = RandomForestClassifier(numTrees=100) # at a certain point the gain in accuracy for more trees will be minimal
gbt_classifier = GBTClassifier() # gradient boosted tree classifier
```

```
dt_model = dt_classifier.fit(train_data)
rf_model = rf_classifier.fit(train_data)
gbt_model = gbt_classifier.fit(train_data)
```

```
dt_prediction = dt_model.transform(test_data)
rf_prediction = rf_model.transform(test_data)
gbt_prediction = gbt_model.transform(test_data)
```

```
# use evaluators
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
```

```
acc_eval = MulticlassClassificationEvaluator(metricName='accuracy')
```

```
print('Decision Tree Accuracy:', acc_eval.evaluate(dt_prediction))
print('Random Forest Accuracy:', acc_eval.evaluate(rf_prediction))
print('Gradient Boosted Accuracy:', acc_eval.evaluate(gbt_prediction))
```
