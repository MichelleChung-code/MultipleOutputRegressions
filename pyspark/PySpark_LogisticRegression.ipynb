```
data = sqlContext.sql('select * from sample_libsvm_data')

# from https://github.com/apache/spark/blob/master/data/mllib/sample_libsvm_data.txt
```

```
from pyspark.ml.classification import LogisticRegression
```

```
display(data.take(5))
```

```
logreg_model = LogisticRegression()
logreg_model = logreg_model.fit(data)
```

```
results = logreg_model.summary
```

```
results.predictions.printSchema()
```

```
display(results.predictions.take(5))
```

```
# create a train-test split
train_data, test_data = data.randomSplit([0.7, 0.3])
final_model = LogisticRegression()
final_model = final_model.fit(train_data)
```

```
prediction_labels = final_model.evaluate(test_data)
```

```
display(prediction_labels.predictions.take(5))
```

```
# evaluation object

from pyspark.ml.evaluation import BinaryClassificationEvaluator, MulticlassClassificationEvaluator
```

```
evaluator = BinaryClassificationEvaluator()
```

```
roc_res = evaluator.evaluate(prediction_labels.predictions)
```

```
roc_res # roc of 1.0 means perfect fit (matched predictions perfectly )
```
