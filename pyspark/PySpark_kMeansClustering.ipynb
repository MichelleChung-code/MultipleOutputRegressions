```
# error decreases as k increases

from pyspark.ml.clustering import KMeans
```

```
# data from https://github.com/apache/spark/blob/master/data/mllib/sample_kmeans_data.txt

data = sqlContext.sql('select * from sample_kmeans_data').select('features')
```

```
k_means = KMeans().setK(3).setSeed(1) # seed number for getting same random number for reproducability
```

```
model = k_means.fit(data)
```

```
# within set sum squared of errors

wssse = model.summary.trainingCost
wssse
```

```
display(data)
```

```
centers = model.clusterCenters()
centers
```

```
res = model.transform(data)
```
